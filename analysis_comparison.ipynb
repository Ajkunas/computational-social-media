# Sentiments

```python
#load data
import pandas as pd

df_f = pd.read_csv('datasets/sentiment_first.csv')
df_l = pd.read_csv('datasets/sentiment_last.csv')
```

```python
fneg = df_f[df_f.label == 'negative']
fpos = df_f[df_f.label == 'positive']
fntl = df_f[df_f.label == 'neutral']
```

```python
lneg = df_l[df_l.label == 'negative']
lpos = df_l[df_l.label == 'positive']
lntl = df_l[df_l.label == 'neutral']
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = [[len(fntl), len(lntl)], #neutral
[len(fpos), len(lpos)], #positiv
[len(fneg), len(lneg)]] #negativ
X = np.arange(2)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(X - 0.25, data[0], color = '#d7f7ff', width = 0.25)
ax.bar(X , data[1], color = '#9acfc9', width = 0.25)
ax.bar(X + 0.25, data[2], color = '#39827a', width = 0.25)
ax.set_ylabel('Numbe of tweets')
ax.set_title('First month and last month sentiment analysis')
ax.set_xticks(X, ('first month', 'last month'))
ax.legend(labels=['Neutral', 'Positive', 'Negative'])
plt.savefig('sentiment_comparison.png', bbox_inches='tight')
plt.show()
```

```python
df_lpos = df_lpos.groupby("emotion").size().reset_index(name="count")
df_lneg = df_lneg.groupby("emotion").size().reset_index(name="count")
```

```python
df_neg_emotions = df_fneg.merge(df_lneg, left_on="emotion", right_on="emotion")
df_neg_emotions["count"] = df_neg_emotions["count_x"] + df_neg_emotions["count_y"]
df_neg_emotions = df_neg_emotions[["emotion", "count"]].sort_values("count", ascending=False).head(10)["emotion"].values.tolist()
df_neg_emotions
```

```python
df_pos_emotions = df_fpos.merge(df_lpos, left_on="emotion", right_on="emotion")
df_pos_emotions["count"] = df_pos_emotions["count_x"] + df_pos_emotions["count_y"]
df_pos_emotions = df_pos_emotions[["emotion", "count"]].sort_values("count", ascending=False).head(10)["emotion"].values.tolist()
df_pos_emotions
```

```python
df_fneg = df_fneg[df_fneg.emotion.isin(df_neg_emotions)]
df_lneg = df_lneg[df_lneg.emotion.isin(df_neg_emotions)]

df_fpos = df_fpos[df_fpos.emotion.isin(df_pos_emotions)]
df_lpos = df_lpos[df_lpos.emotion.isin(df_pos_emotions)]
```

```python
df_fneg["month"] = "first"
df_lneg["month"] = "last"
df_fpos["month"] = "first"
df_lpos["month"] = "last"
```

```python
neg_frames = [df_fneg, df_lneg]
df_neg = pd.concat(neg_frames)

pos_frames = [df_fpos, df_lpos]
df_pos = pd.concat(pos_frames)
```

```python
# plot top 5 positive topics as horizontal bar chart
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(7, 5))
sns.barplot(data=df_neg, x="count", y="emotion", hue="month", palette=["#1b9df0", "#77CAFF"])
plt.title('Top 5 negative emotions')
plt.xlabel('Count')
plt.ylabel('Emotion')
plt.savefig('emotion_neg_comparison.png', bbox_inches='tight')
plt.show()
```
